```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Inner Join

```python
left_table = {'Date': ['01-01-2020', '02-01-2020', '03-01-2020', '04-01-2020' ],
              'CountryID': [1,1,3,2],
              'Units': [40, 25, 30, 35]}

left_table_df = pd.DataFrame(left_table)
left_table_df
```

```python
right_table = {'ID': [3,4],
              'Country': ['Pandama', 'Spain']}
right_table_df = pd.DataFrame(right_table)
right_table_df
```

```python
left_table_df.merge(right_table_df, left_on="CountryID", right_on="ID")[["ID","Country"]]
```

### Semi Join
###### A semi join is a type of join in relational databases that combines rows from two tables based on a related column but only returns rows from one of the tables. Unlike traditional joins, a semi join does not include columns from the second table in the output. It is primarily used to filter rows in one table based on matching values in another.

```python
left_table_df.merge(right_table_df, left_on="CountryID", right_on="ID")
```

### Left Join

```python
left_table = {'Date': ['01-01-2020', '02-01-2020', '03-01-2020', '04-01-2020' ],
              'CountryID': [1,1,3,4],
              'Units': [40, 25, 30, 35]}

left_table_df = pd.DataFrame(left_table)
left_table_df

right_table = {'ID': [1,2, 3],
              'Country': ['USA', 'Canada', 'Panama']}

right_table_df = pd.DataFrame(right_table)
right_table_df

# Perform a left join
result = pd.merge(left_table_df, right_table_df, how='left', left_on='CountryID', right_on='ID')

# Display the result
print(result)
```

```python
result_2 = pd.merge(left_table_df, right_table_df, how="right", left_on="CountryID", right_on="ID")
result_2
```

```python
result_3 = pd.merge(left_table_df, right_table_df, how="outer", left_on="CountryID", right_on="ID")
result_3
```
